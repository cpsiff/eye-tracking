```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

DATA_FILE = "DATA_FINAL.csv"
STYLES = "bionic", "control", "random"

# read data into dataframe
data = pd.read_csv(DATA_FILE)
data
```

```python
# get the average value of each column
cols = [col for col in data.columns if "norm" not in col and col not in ['index', 'user', 'gender', 'text_order', 'style_order', 'occupation', 'preference', 'response']]
means = data[cols].mean()
means
```

```python
# mean plots
plt.style.use("fivethirtyeight")

# Average number of fixations
fig, ax = plt.subplots()
stats = [f"num_fixations_{style}" for style in STYLES]
ax.bar([0, 1, 2], [means[stat] for stat in stats])
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["Bionic", "Control", "Random"])
ax.set_ylabel("Average number of fixations")
ax.set_xlabel("Text style")

# Average number of saccades
fig, ax = plt.subplots()
stats = [f"num_saccades_{style}" for style in STYLES]
ax.bar([0, 1, 2], [means[stat] for stat in stats])
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["Bionic", "Control", "Random"])
ax.set_ylabel("Average number of saccades")
ax.set_xlabel("Text style")

# Average fixation length
fig, ax = plt.subplots()
stats = [f"avg_fixation_length_{style}" for style in STYLES]
ax.bar([0, 1, 2], [means[stat]/1000 for stat in stats]) # divide by 1000 to convert microseconds to milliseconds
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["Bionic", "Control", "Random"])
ax.set_ylabel("Average fixation length (ms)")
ax.set_xlabel("Text style")

# Average saccade length
fig, ax = plt.subplots()
stats = [f"avg_saccade323432321`_length_{style}" for style in STYLES]
ax.bar([0, 1, 2], [means[stat]/1000 for stat in stats]) # divide by 1000 to convert microseconds to milliseconds
ax.set_xticks([0, 1, 2])
ax.set_xticklabels(["Bionic", "Control", "Random"])
ax.set_ylabel("Average fixation length (ms)")
ax.set_xlabel("Text style")
```
